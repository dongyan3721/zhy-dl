```python
w = torch.tensor([1], requires_grad=True, device='mps', dtype=torch.float32)
x = torch.tensor([2],  requires_grad=True, device='mps', dtype=torch.float32)

a = w+x
b = w+1
y = a @ b
y.backward()
print(w.grad)
```

```python
class OneDimGradTest(torch.autograd.Function):
    @staticmethod
    def forward(ctx, x):
        y = 2.5 * torch.pow(x, 5) + torch.pow(x, 2) + x
        ctx.save_for_backward(x)
        return y

    @staticmethod
    def backward(ctx, grad_output):
        x, = ctx.saved_tensors
        return grad_output * (12.5 * torch.pow(x, 4) + 2*x + 1)

x = torch.tensor([2], requires_grad=True, device='mps', dtype=torch.float)
y = OneDimGradTest.apply(x)
y.backward()
print(x.grad)
```
