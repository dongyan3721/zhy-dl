## 数据集


```python
from torch.utils.data import Dataset, DataLoader
import pandas as pd
from sklearn.model_selection import train_test_split

movie_file_root = './dataset/ratings.csv'
df = pd.read_csv(movie_file_root)

df_t, df_v = train_test_split(df, test_size=0.2)


class MovieLensDataset(Dataset):
    def __init__(self, mv_df, transform=None):
        self.transform = transform
        self.users = mv_df['userId'].values
        self.movies = mv_df['movieId'].values
        self.ratings = mv_df['rating'].values

    def __len__(self):
        return len(self.users)

    def __getitem__(self, idx):
        return self.users[idx], self.movies[idx], self.ratings[idx]

test_data_loader = DataLoader(MovieLensDataset(df_t), batch_size=512, shuffle=True)
validate_data_loader = DataLoader(MovieLensDataset(df_v), batch_size=512, shuffle=False)

for user, movie, rating in test_data_loader:
    print(user, movie, rating)
```
